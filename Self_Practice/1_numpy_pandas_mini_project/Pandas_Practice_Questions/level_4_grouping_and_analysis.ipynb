## 16 group students by Age and calculate Average marks 

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['John Doe', 'Jane Smith', 'Mike Johnson', 'Emily Brown', 'David Wilson','Sanjeev Khadka','Sulav Khadka','sanjay kumar '],
    'Age': [20, 21, 22, 23, 24,20,21,22],
    'Marks': [85, 92, 78, 90, 88,99,80,70]
}

df = pd.DataFrame(data)

result = df.groupby("Age")["Marks"].mean()
print(result)
```

## 17 Find the student with highest marks.


```python
data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['John Doe', 'Jane Smith', 'Mike Johnson', 'Emily Brown', 'David Wilson','Sanjeev Khadka','Sulav Khadka','sanjay kumar '],
    'Age': [20, 21, 22, 23, 24,20,21,22],
    'Marks': [85, 92, 78, 90, 88,99,80,70]
}

df2 = pd.DataFrame(data)


highest_marks = df2["Marks"].max()
print(highest_marks)

result  = df2[df2["Marks"] == highest_marks]["Name"]
print(result)


```

## 18. Count how many students passed.


```python
data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['John Doe', 'Jane Smith', 'Mike Johnson', 'Emily Brown', 'David Wilson','Sanjeev Khadka','Sulav Khadka','sanjay kumar '],
    'Age': [20, 21, 22, 23, 24,20,21,22],
    'Marks': [40, 92, 78, 90, 88,99,80,70]
}

df3 = pd.DataFrame(data)
print(df3)

result = df3[df3["Marks"] >= 50]
print("Number of Students passed is " , len(result))
```

## 19 Apply a custom function to increase marks by 5%.

```python
data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['John Doe', 'Jane Smith', 'Mike Johnson', 'Emily Brown', 'David Wilson','Sanjeev Khadka','Sulav Khadka','sanjay kumar '],
    'Age': [20, 21, 22, 23, 24,20,21,22],
    'Marks': [40, 92, 78, 90, 88,99,80,70]
}

df4 = pd.DataFrame(data)
print(df4)


df4["Marks"] = df4["Marks"].apply(lambda mark : (5/100) * mark + mark )
print(df4)
```
