## 6 Select only the Name and Marks columns.


```python
dict = {
    "Name":["Sanjeev","Deepa","Samir","Sulav","Jaidev"],
    "Age":[10,20,30,40,90],
    "Marks":[90,80,70,95,10]
}
df1 = pd.DataFrame(dict)
print(df1)
print()
print(df1["Name"])
print(df1["Marks"])
```

## 7 Filter students whose Marks > 80.

```python
filtered_students = df1[df1["Marks"] > 80]
print(filtered_students)
```

## 8 Filter rows where Age is between 18 and 22 

```python

filtered_students = df1[(df1["Age"] > 18 ) & (df1["Age"] < 22 )]
print(filtered_students)
```

## 9 Sort the DataFrame by Marks (descending).

```python
print(df1.sort_values('Marks',ascending=False))
```

## 10 Add a new column “Passed” where Marks ≥ 40.

```python
# df1["passed"] = df1[df1["Marks"] >= 40]
# print(df1)
df1["passed"]  = df1[df1["Marks"] >= 40]["Name"]
print(df1)
```
