##  1 Create a DataFrame using a dictionary:


```python
# Name, Age,Marks

dict = {
    "name": "sanjeev khadka",
    "cls": "12",
    "Age": "18",
}

# df = pd.DataFrame(dict)
# print(df)

#! Note , It Gives This error  =>
# * Summary: Pandas DataFrame Scalar Error
# Error You Got
# ValueError: If using all scalar values, you must pass an index

# Cause

# When creating a DataFrame from a dictionary:

# All values are scalars (single values) → Pandas doesn’t know how many rows to create.
# ~ Pandas requires either lists/arrays to define multiple rows or an index to define the number of rows.


#! What is meant By Scalar ?
# => A scalar is simply a single value, not a list, array, or collection of values.
# In programming and math:
# Scalar = one single value
# Vector / Array / List = multiple values
```

```python
# Solutions For that 

# 1 solution Wrap scalars in lists
df1 = pd.DataFrame({
    "name":["sanjeev khadka "],
    "cls" : ["12"],
    "Age":["18"]
})
print(" \n ")
print(df1)


# 2 solution Provide an explicit index 
df2 = pd.DataFrame({
    "name": "sanjeev khadka",
    "cls": "12",
    "Age": "18"
}, index=[0])
print(" \n ")
print(df2)

# 3 Use lists to define multiple rows
df3 = pd.DataFrame({
    "name": ["sanjeev", "ram", "hari", "sita"],
    "cls": ["12", "12", "12", "12"],
    "Age": [18, 19, 18, 20],
    "marks": [10, 20, 30, 40]
})
print(" \n ")
print(df3)

```

## 5 Set one column as index.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# setting column to 'A' as the index 

df_set_index = df.set_index('A')
print(df_set_index)
```
