## 11  Replace missing values with the mean.


```python

data = {
    'Employee_ID': [1, 2, 3, 4, 5],
    'Name': ['John Doe', 'Jane Smith', 'Sanjeev_KhadkaX', 'Emily Brown', 'David Wilson'],
    'Department': ['Sales', 'Marketing', 'CEO', 'HR', 'Finance'],
    'Salary': [50000, None, 10000000, 8000, 9000]
}

df = pd.DataFrame(data)
print(df)

result = df.fillna(df['Salary'].mean())


print(result)
```

## 12 Drop rows with missing values.

```python
print(df)
result = df.dropna()
```

## 13 Check How many null values each column has 

```python
print(df)
result = df.isna().sum()
print(result)
```

## 14 Convert a column datatype 

```python
print(df.dtypes)
df.fillna(df['Salary'].mean(),inplace=True)
df['Salary'] = df['Salary'].astype(int)
print(df.dtypes)
```

## 15 Remove Duplicate Rows 

```python

data = {
    'Employee_ID': [1, 2, 2, 3, 3, 3],
    'Name': ['John Doe', 'Jane Smith', 'Jane Smith', 'Mike Johnson', 'Mike Johnson', 'Mike Johnson'],
    'Department': ['Sales', 'Marketing', 'Marketing', 'Engineering', 'Engineering', 'Engineering'],
    'Salary': [50000, 60000, 60000, 70000, 70000, 70000]
}

df = pd.DataFrame(data)
print(df)
print()

# Drop duplicate rows based on all columns
df_without_duplicates = df.drop_duplicates()

print(df_without_duplicates)
```
