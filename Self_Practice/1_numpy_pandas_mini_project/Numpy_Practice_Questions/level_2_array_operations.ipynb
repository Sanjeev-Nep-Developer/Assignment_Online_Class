```python
# 6 Create two arrays of size 5 and perform 
# addition
# subtraction
# multiplication 

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,20,30,40,50])

print(arr1+ arr2)
print(arr1-arr2)
print(arr1 * arr2)
```

```python
# 7 Multiply every element of an array by 3.
print(arr1)
result = arr1 * 3 
print(result)
```

```python
# 8 Find the maximum, minimum, mean, and standard deviation of an array.
arr3 = np.array(np.arange(20).reshape(5,4))
print(type(arr3))
print(np.mean(arr3))
print(np.std(arr3))
print(np.max(arr3))
```

```python
# 9 Replace all odd numbers in an array with -1.

odd_indices = arr3 % 2 != 0
print(odd_indices)

arr3[odd_indices] = -1
print(arr3)
```

```python
# 10 Reshape a 1D array of size 12 into a 3×4 matrix.

arr4 = np.arange(12)
print(arr4)
print(arr4.ndim)
print(arr4.size)

new_arr4 = arr4.reshape(3,4)
print(new_arr4)
print(new_arr4.size)
```
