```python
# 1 Create a Numpy array of integers from 10 to 50 

numpy_array1 = np.array(np.arange(10,51))
print(numpy_array1)
print(type(numpy_array1))
```

```python
# 2 Create a 5×5 matrix filled with zeros.
numpy_array2 = np.zeros((5,5))
print(numpy_array2)
```

```python
# 3 Create a 3×3 identity matrix.
numpy_array3 = np.identity(3)
# numpy_array3 = np.identity((3,3)) # {This gives error because , .idenitiy() expects only one number as ,it always creates square matrix so why to give two values one value is enough , so it is designed in such a way where it only accepts one integer.. }

print(numpy_array3)
print(numpy_array3.shape)
```

```python
# 4 Generate an array of 20 random numbers between 0 and 1.

numpy_array4 = np.random.rand(20)
print(numpy_array4)
print(numpy_array4.size)
print(numpy_array4.shape)
```

```python
# 5 Find the shape, size, and datatype of a given array.
numpy_array5 = np.random.randint(0,20,20).reshape(5,4)
print(numpy_array5)
print(numpy_array5.size)
print(numpy_array5.shape)
print(numpy_array5.dtype)
```
