```python
# 11 Extract the 2nd row from a 2D array.

arr1 = np.arange(20).reshape(5,4)
print(arr1)


print(arr1[1])
```

```python
# 12 Extract the last column from a matrix 
last_column = arr1[:,-1]
print(last_column)
```

```python
# 13 From a 1D array, select all values greater than 25.

arr_1D = np.random.randint(15,50,20)
print(arr_1D)
print(arr_1D.size)

result = arr_1D[arr_1D > 25]

print("All values greater than 25 are : ",result)

```

```python
# 14 Modify a slice of an array without using a loop.

arr = np.array(np.arange(20))
print(arr)
print(arr.size)

arr[4:8] = 10
print(arr)
```

```python
# 15 Reverse a NumPy array.

print(arr)

result = arr[::-1]
print(result)
```
