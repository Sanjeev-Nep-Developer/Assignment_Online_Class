```python
# 16 Find indices where values are greater than the mean .. 
values = np.random.randint(20,60,10)
print(values)

mean_value = np.mean(values)
print(mean_value)

indices = values > mean_value
print(indices)
```

```python
# 17 Check if two arrays are equal element-wise.
arr_1 = np.array([10,20,30,40,50])
arr_2 = np.array([1,2,3,4,5])


result = np.array_equal(arr_1,arr_2)

print(result)
```

```python
# 18 Sort an array.
numpy_array = np.array([70,80,90,10,20,30,40,50])
result = np.sort(numpy_array)
print(result)
```

```python
# 19 Compute the dot product of two arrays.
arr_1 = np.array([10,20,30,40,50])
arr_2 = np.array([1,2,3,4,5])

result = np.dot(arr_1,arr_2)
print(result)
```

```python
# 20 Stack two arrays:
# Vertically
arr_1 = np.array([10,20,30,40,50])
arr_2 = np.array([1,2,3,4,5])

result1 = np.vstack((arr_1,arr_2))
print(result1)

# Horizontally
result2 = np.hstack((arr_1,arr_2))
print(result2)
```
