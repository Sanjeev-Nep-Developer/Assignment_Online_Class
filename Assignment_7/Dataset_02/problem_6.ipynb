```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Problem 6: Aggregation and Transformation 
# 1 Perform a groupby operation on Neighborhood and calculate the mean and median for SalePrice.
# 2 Use the apply() function to create a custom transformation for SalePrice (e.g., scaling or transformation).
# 3 Aggregate multiple statistical values (e.g., mean, median, and count) for SalePrice grouped by OverallQual.
```

```python
df = pd.read_csv('housing.csv')
```

```python
# 1 Perform a groupby operation on Neighborhood and calculate the mean and median for SalePrice.
result = df.groupby('Neighborhood')['SalePrice'].agg(['mean','median'])
print(result)
```

```python
# 2 Use the apply() function to create a custom transformation for SalePrice (e.g., scaling or transformation).
# Min-max scaling: (value - min) / (max - min)
result = df['SalePrice'].apply(lambda x: (x - df['SalePrice'].min()) / (df['SalePrice'].max() - df['SalePrice'].min()))

print(result)
```

```python
# 3 Aggregate multiple statistical values (e.g., mean, median, and count) for SalePrice grouped by OverallQual.

result = df.groupby("OverallQual")["SalePrice"].agg(["mean","median","count"])
print(result)
```
