```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

### 1 Load the House Prices dataset using Pandas and perform initial exploration:

### Display the first and last 5 rows.
### Display the data types of each column.
### Check for missing values and calculate the percentage of missing data for each column.
### Summarize the statistical properties of the numerical columns using describe().

```python
df = pd.read_csv("housing.csv")
```

```python
#  Display the first and last 5 rows.
print(df.head())
print(df.tail())
```

```python
#  Display the data types of each column.
print(df.dtypes)
```

```python
# Check for missing values and calculate the percentage of missing data for each column.
result = df.isna().sum()
print(result)

number_of_total_values_in_total_bedrooms = len(df["total_bedrooms"])
print(number_of_total_values_in_total_bedrooms)
missing_values_of_total_bedrooms = result["total_bedrooms"]
print(missing_values_of_total_bedrooms)


# percentage of missing data .. 
calc = (missing_values_of_total_bedrooms/number_of_total_values_in_total_bedrooms) * 100
print(calc)
```

```python
# Summarize the statistical properties of the numerical columns using describe().
print(df.describe())
```
