```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("housing.csv")
print(df.dtypes)
```

###  1 Convert the numerical columns (such as LotArea, YearBuilt, GrLivArea, SalePrice, etc.) into a NumPy array.


```python
lotAreaValues = df["LotArea"]
YearBuiltValues = df["YearBuilt"]
GrLivAreaValues = df["GrLivArea"]
SalePriceValues = df["SalePrice"]


print(type(lotAreaValues))
print(YearBuiltValues)
print(GrLivAreaValues)
print(SalePriceValues)
```

```python
np1 = np.array(lotAreaValues) 
np2 =  np.array(YearBuiltValues)
np3 =  np.array(GrLivAreaValues)
np4 =  np.array(SalePriceValues)


print(type(np1))
print(type(np2))
print(type(np3))
print(type(np4))


print(np1[0:20],end="\n")
print(np2[0:20],end="\n")
print(np3[0:20],end="\n")
print(np4[0:20],end="\n")
```

### 2 Perform the following NumPy operations on the array:
### - Compute the mean and standard deviation of each column.
### - Normalize the data using min-max scaling and z-score normalization.
### - Compute the dot product between two numerical columns of your choice.

```python
mean_np1 = np.nanmean(np1)
mean_np2 = np.nanmean(np2)
mean_np3 = np.nanmean(np3)
mean_np4 = np.nanmean(np4)

print(mean_np1,mean_np2,mean_np3,mean_np4)
```

```python
std_np1 = np.nanstd(np1)
std_np2 = np.nanstd(np2)
std_np3 = np.nanstd(np3)
std_np4 = np.nanstd(np4)


print(std_np1,std_np2,std_np3,std_np4)
```

```python
#- Normalize the data using min-max scaling and z-score normalization.


# Convert numerical columns to numpy array
numerical_cols = ['SalePrice', 'GrLivArea', 'LotArea', 'YearBuilt', 'OverallQual']
arr = df[numerical_cols].to_numpy()

# Min-Max scaling
arr_minmax = (arr - np.nanmin(arr, axis=0)) / (np.nanmax(arr, axis=0) - np.nanmin(arr, axis=0))

# Z-score normalization
arr_zscore = (arr - np.nanmean(arr, axis=0)) / np.nanstd(arr, axis=0)
```
