```python
# Problem 3: Handling Missing Data and Imputation
# 1. Identify columns with missing values.
# 2. Impute the missing values:
#  - For numerical columns, replace missing values with the median.
# -  For categorical columns, replace missing values with the most frequent value.
# Justify your choice of imputation methods and check if any missing data remains.
```

```python
df = pd.read_csv('housing.csv')
```

```python
# 1. Identify columns with missing values.
print(df.isna().sum())
print(df.dtypes)
```

```python
# 2. Impute the missing values:
#  - For numerical columns, replace missing values with the median.
# -  For categorical columns, replace missing values with the most frequent value.

numerical_columns = df.select_dtypes(include='number')
categorical_columns = df.select_dtypes(include='object')

print("Numerical columns are \n" , numerical_columns.head(),end="\n\n")
print("Categorical columns are \n" , categorical_columns.head(),end="\n\n")

missing_value_of_numerical_columns = numerical_columns.isna().sum()
missing_value_of_categorical_columns = categorical_columns.isna().sum()


print("Missing Values of Numerical are \n" , missing_value_of_numerical_columns,end="\n\n")
print("Missing values of Categorical  are \n",missing_value_of_categorical_columns,end="\n\n")


# Filling the missing values of numerical columns
numerical_columns["GrLivArea"] = numerical_columns["GrLivArea"].fillna(df["GrLivArea"].mean())
numerical_columns["LotArea"] = numerical_columns["LotArea"].fillna(df["LotArea"].mean())

# Filling the missing values of Categorical columns 
categorical_columns["Neighborhood"] = categorical_columns["Neighborhood"].mode()[0]


```

```python
missing_value_of_numerical_columns = numerical_columns.isna().sum()
print(missing_value_of_numerical_columns)
```

```python
missing_value_of_categorical_columns = categorical_columns.isna().sum()
print("Missing values of Categorical  are \n",missing_value_of_categorical_columns,end="\n\n")
```
